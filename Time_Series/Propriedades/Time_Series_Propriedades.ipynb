# Propriedades das séries temporais

Objetivo é obter as componentes da tendência, sazonalidade e ruído da série temporal

Utilizaremos uma série temporal da temperatura na cidade australiana de Melbourne

* Importando biblitecas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

* Lendo arquivo CSV

```python
dados = pd.read_csv('daily-min-temperatures.csv')
```

* Exibindo cinco primeiras linhas

```python
dados.head()
```

* Exibindo informação do dado

```python
dados.info()
```

* Convertendo coluna date para DateTime object

```python
dados['Date'] = pd.to_datetime(dados['Date'])
```

* Graficando série temporal

```python
dados = dados.set_index('Date')
```

```python
dados.head()
```

```python
dados.plot()
plt.xticks(rotation=45)
plt.xlabel('Data')
plt.ylabel('Temperatura')
plt.tight_layout()
```

* Fazendo decomposição sazonal

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
resultado = seasonal_decompose(dados['Temp'], period=12)
```

* Obtendo componentes

```python
tendencia = resultado.trend
sazonalidade = resultado.seasonal
residuo = resultado.resid
```

* Graficando componentes

```python
plt.plot(tendencia)
plt.xlabel('Data')
plt.ylabel('Temperatura')
plt.title('Tendência')
```

```python
plt.plot(sazonalidade)
plt.xlabel('Data')
plt.ylabel('Temperatura')
plt.title('Sazonalidade')
```

```python
plt.plot(residuo)
plt.xlabel('Data')
plt.ylabel('Temperatura')
plt.title('Resíduo')
```
