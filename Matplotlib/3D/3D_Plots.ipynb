```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
```

```python
x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)
```

```python
X,Y = np.meshgrid(x,y)
```

```python
Z1 = 8*X+3*Y-10
```

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(111, projection='3d')
mycmap = plt.get_cmap('rainbow')
surf1 = ax.plot_surface(X, Y, Z1, cmap=mycmap)
fig.colorbar(surf1, ax=ax, shrink=0.5, aspect=5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Z=F(x,y)')
plt.tight_layout()
```

```python
Z2 = -5*X+3*Y+2*X*Y-5
```

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(111, projection='3d')
mycmap = plt.get_cmap('rainbow')
surf1 = ax.plot_surface(X, Y, Z2, cmap=mycmap)
fig.colorbar(surf1, ax=ax, shrink=0.5, aspect=5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Z=F(x,y)')
plt.tight_layout()
```
