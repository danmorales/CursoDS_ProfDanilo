* Importando bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

* Criando variável X com números aleatórios

```python
X = np.random.rand(2000)
```

* Criando histograma

```python
plt.hist(X)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.show()
```

* Alterando número de bins

```python
plt.hist(X,bins=10)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=10')
plt.show()
```

```python
plt.hist(X,bins=20)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=20')
plt.show()
```

```python
plt.hist(X,bins=30)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=30')
plt.show()
```

* Alterando cores

```python
plt.hist(X,bins=50,color='red')
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=50')
plt.show()
```

```python
plt.hist(X,bins=50,color='red', edgecolor='blue')
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=50')
plt.show()
```

* Multiplos histogramas na mesma imagem

```python
X1 = np.random.normal(0,1,size = 800)
X2 = np.random.normal(3,1,size = 800)
X3 = np.random.normal(5,1,size = 800)
```

```python
plt.hist(X1,bins=10,color='red')
plt.hist(X2,bins=10,color='blue')
plt.hist(X3,bins=10,color='green')
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Multiplos histogramas')
plt.show()
```

* Distribuição cumulativa

```python
X = np.random.rand(2000)
```

```python
plt.hist(X,bins=20,color='red', edgecolor='blue',cumulative=True)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Histograma cumulativo')
plt.show()
```
